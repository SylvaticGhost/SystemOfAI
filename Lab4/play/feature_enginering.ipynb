```python
EMPTY_COLOR = np.array([0, 0, 0], dtype=np.uint8)
WALL_COLOR = np.array([84, 92, 214], dtype=np.uint8)
ENEMY_COLOR = np.array([210, 210, 64], dtype=np.uint8)
PLAYER_COLOR = np.array([240, 170, 103], dtype=np.uint8)

EMPTY_INDEX = 0
WALL_INDEX = 1
ENEMY_INDEX = 2
PLAYER_INDEX = 3

PORTAL_INDEX = 4 # add manualy
PORTAL_COLOR = np.array([74, 255, 56], dtype=np.uint8)
```

```python
replaced_state = np.zeros(frame.shape[:2], dtype=np.uint8)
mask_player = np.all(frame == PLAYER_COLOR, axis=2)
mask_wall = np.all(frame == WALL_COLOR, axis=2)
mask_enemy = np.all(frame == ENEMY_COLOR, axis=2)
replaced_state[mask_player] = PLAYER_INDEX
replaced_state[mask_wall] = WALL_INDEX
replaced_state[mask_enemy] = ENEMY_INDEX
replaced_state
```

```python
skip_layers = 0
while True:
    if replaced_state[skip_layers][skip_layers] == 0:
        skip_layers += 1
    else:
        break
print(f'Skipped {skip_layers} layers of empty pixels')
replaced_state = replaced_state[skip_layers:-skip_layers, skip_layers:-skip_layers]
print(f'New shape: {replaced_state.shape}')
```

```python
skip_layers_from_bottom = 0
while True:
    if replaced_state[-(skip_layers_from_bottom + 1)][-(skip_layers_from_bottom + 1)] == 0:
        skip_layers_from_bottom += 1
    else:
        break
print(f'Skipped {skip_layers_from_bottom} layers of empty pixels from bottom')
replaced_state = replaced_state[:-skip_layers_from_bottom, :]
print(f'New shape: {replaced_state.shape}')
```

```python
def fill_part(matrix, a, b, c, d, fill_value):
    rows, cols = matrix.shape
    min_i = max(0, min(a[0], b[0], c[0], d[0]))
    max_i = min(rows - 1, max(a[0], b[0], c[0], d[0]))
    min_j = max(0, min(a[1], b[1], c[1], d[1]))
    max_j = min(cols - 1, max(a[1], b[1], c[1], d[1]))

    for ii in range(min_i, max_i + 1):
        for jj in range(min_j, max_j + 1):
            if matrix[ii, jj] == EMPTY_INDEX:
                matrix[ii, jj] = fill_value
```

```python
#Iterate over border to fill holes in wall and find portals
PORTAL_MIN_LENGHT = 3
PORTAL_WIDTH = 0
WALL_WIDTH = 3

i, j = 0, 0
height, width = replaced_state.shape
last_wall_pixel = (0, 0)
while True:
    if replaced_state[i][j] == WALL_INDEX:
        distance_to_last_wall = max(np.abs(i - last_wall_pixel[0]), np.abs(j - last_wall_pixel[1]))
        if distance_to_last_wall == 1:
            pass
        else:
            material_to_fill = WALL_INDEX if distance_to_last_wall < PORTAL_MIN_LENGHT else PORTAL_INDEX
            width_to_fill = WALL_WIDTH if material_to_fill == WALL_INDEX else PORTAL_WIDTH
            A,B,C,D = None, None, None, None
            if j == 0:
                A = (last_wall_pixel[0] + 1, last_wall_pixel[1])
                B = (last_wall_pixel[0] + 1, last_wall_pixel[1] + width_to_fill)
                C = (i, j + width_to_fill)
                D = (i, j)
            elif i == height - 1:
                A = (last_wall_pixel[0], last_wall_pixel[1] + 1)
                B = (last_wall_pixel[0] + width_to_fill, last_wall_pixel[1] + 1)
                C = (i - width_to_fill, j)
                D = (i, j)
            elif j == width - 1:
                A = (last_wall_pixel[0] - 1, last_wall_pixel[1])
                B = (last_wall_pixel[0] - 1, last_wall_pixel[1] - width_to_fill)
                C = (i, j - width_to_fill)
                D = (i, j)
            elif i == 0:
                A = (last_wall_pixel[0], last_wall_pixel[1] - 1)
                B = (last_wall_pixel[0] - width_to_fill, last_wall_pixel[1] - 1)
                C = (i + width_to_fill, j)
                D = (i, j)
            else:
                raise ValueError("Unexpected wall pixel not on border")

            fill_part(replaced_state, A, B, C, D, material_to_fill)

        last_wall_pixel = (i, j)

    # Move to next pixel. Direction: down, right, up, left
    if j == 0 and i < height - 1:
        i += 1
    elif i == height - 1 and j < width - 1:
        j += 1
    elif j == width - 1 and i > 0:
        i -= 1
    elif i == 0 and j > 0:
        j -= 1

    if i == 0 and j == 0:
        break

replaced_state
```

## Ilustractate feature engineering steps

```python
colored_state = np.zeros((*replaced_state.shape, 3), dtype=np.uint8)
colored_state[replaced_state == EMPTY_INDEX] = EMPTY_COLOR
colored_state[replaced_state == WALL_INDEX] = WALL_COLOR
colored_state[replaced_state == ENEMY_INDEX] = ENEMY_COLOR
colored_state[replaced_state == PLAYER_INDEX] = PLAYER_COLOR
colored_state[replaced_state == PORTAL_INDEX] = PORTAL_COLOR
plt.imshow(colored_state)
```

```python
## find player position
player_positions = np.argwhere(replaced_state == PLAYER_INDEX)
i_pos = player_positions[:, 0]
j_pos = player_positions[:, 1]
box = (np.min(i_pos), np.min(j_pos), np.max(i_pos), np.max(j_pos))
print(f'Player bounding box: {box}')
```

```python
ls = np.linspace(box[0], box[2],num=(np.abs(box[2] - box[0]) + 1) // 4, dtype=int)
left_bounce_points = []
for i in ls:
    left_bounce_points.append((i, box[1]))
left_bounce_points
```

```python
us = np.linspace(box[1], box[3],num=(np.abs(box[3] - box[1]) + 1) // 3, dtype=int)
upper_bounce_points = []
for j in us:
    upper_bounce_points.append((box[0], j))
upper_bounce_points
```

```python
right_bounce_points = []
for i in ls:
    right_bounce_points.append((i, box[3]))
right_bounce_points
```

```python
down_bounce_points = []
for j in us:
    down_bounce_points.append((box[2], j))
down_bounce_points
```

```python
left_search_point = ((box[0] + box[2]) // 2, box[1])
right_search_point = ((box[0] + box[2]) // 2, box[3])
up_search_point = (box[0], (box[1] + box[3]) // 2)
down_search_point = (box[2], (box[1] + box[3]) // 2)
```

```python
l_wall = (left_search_point[0], 0)
r_wall = (right_search_point[0], replaced_state.shape[1] - 1)
u_wall = (0, up_search_point[1])
d_wall = (replaced_state.shape[0] - 1, down_search_point[1])

reach_l, reach_r, reach_u, reach_d = False, False, False, False

closest_enemy = None
```

```python
i = 0
while True:
    if not reach_l:
        cord = (left_search_point[0], left_search_point[1] - i)
        material = replaced_state[cord]
        if material == WALL_INDEX:
            reach_l = True
            l_wall = cord
        elif material == ENEMY_INDEX:
            if closest_enemy is None:
                closest_enemy = cord
    if not reach_r:
        cord = (right_search_point[0], right_search_point[1] + i)
        material = replaced_state[cord]
        if material == WALL_INDEX:
            reach_r = True
            r_wall = cord
        elif material == ENEMY_INDEX:
            if closest_enemy is None:
                closest_enemy = cord
    if not reach_u:
        cord = (up_search_point[0] - i, up_search_point[1])
        material = replaced_state[cord]
        if material == WALL_INDEX:
            reach_u = True
            u_wall = cord
        elif material == ENEMY_INDEX:
            if closest_enemy is None:
                closest_enemy = cord
    if not reach_d:
        cord = (down_search_point[0] + i, down_search_point[1])
        material = replaced_state[cord]
        if material == WALL_INDEX:
            reach_d = True
            d_wall = cord
        elif material == ENEMY_INDEX:
            if closest_enemy is None:
                closest_enemy = cord
    if reach_l and reach_r and reach_u and reach_d:
        break
    i += 1

print(f'Left wall: {l_wall}, Right wall: {r_wall}, Up wall: {u_wall}, Down wall: {d_wall}')
print(f'Closest enemy: {closest_enemy}')
```
